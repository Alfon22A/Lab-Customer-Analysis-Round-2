### Dealing with the data

1. Show the dataframe shape.

```python
# First we load the file

import pandas as pd

df1 = pd.read_excel("files_for_lab/excel_files/marketing_customer_analysis.xlsx")
df1.shape
```

2. Standardize header names.

```python
df1.head()
```

```python
# We drop the first column

df1 = df1.drop(["Unnamed: 0"], axis = 1)
df1
```

```python
# Now we recycle functions that we defined in the previous lab standarize the labels

def standard_col(x):
    cols = []
    for item in x:
        col = ""
        for letter in item:
            if letter == " ":
                letter = "_"
                col += letter
            else:
                col += letter
        cols.append(col.lower())
    return cols

df1.columns = standard_col(df1.columns)

df1.head()
```

```python
# We fix the last column that is not standard

df1 = df1.rename(columns={"employmentstatus":"employment_status"})

df1.columns
```

5. Check and deal with `NaN` values.

```python
df1.isna().sum()
```

```python
display(df1["state"].value_counts(dropna=False))
print("--")
display(df1["response"].value_counts(dropna=False))
print("--")
display(df1["months_since_last_claim"].value_counts(dropna=False))
print("--")
display(df1["number_of_open_complaints"].value_counts(dropna=False))
print("--")
display(df1["vehicle_class"].value_counts(dropna=False))
print("--")
display(df1["vehicle_size"].value_counts(dropna=False))
print("--")
display(df1["vehicle_type"].value_counts(dropna=False))
```

```python
display(df1[df1['state'].isna()==True].index)
display(df1[df1['response'].isna()==True].index)
display(df1[df1['months_since_last_claim'].isna()==True].index)
display(df1[df1['number_of_open_complaints'].isna()==True].index)
display(df1[df1['vehicle_class'].isna()==True].index)
display(df1[df1['vehicle_size'].isna()==True].index)
# Looks like they are pairs of columns with the same missing rows
```

```python
# We are probable better off dropping the state and response columns as they are not that many and it is impossible to complete (response could be No as default)

df1 = df1[df1['state'].isna()==False]

df1.isna().sum()
```

```python
# In the case of months_since_last_claim and number_of_open_complaints, we can assume that the value is 0 as a default

df1['months_since_last_claim'] = df1['months_since_last_claim'].fillna(0)
df1['number_of_open_complaints'] = df1['number_of_open_complaints'].fillna(0)

df1.isna().sum()
```

```python
# For the vehicle_class and vehicle_size we can either drop them or assign the mode

df1['vehicle_class'] = df1['vehicle_class'].fillna(df1['vehicle_class'].mode()[0], inplace=True)
df1['vehicle_size'] = df1['vehicle_size'].fillna(df1['vehicle_size'].mode()[0], inplace=True)

df1.isna().sum()
```

```python
# For the last column with missing values, dropping the rows is out of the question as it is half our dataset, so we will assign the only value that exists or drop the column altogether

df1 = df1.drop(["vehicle_type"], axis=1)
df1
```

```python
df1.select_dtypes(np.number)
```

```python
# Some columns make no sense as floats

df1 = df1.astype({"income":'int',"monthly_premium_auto":'int', "months_since_last_claim":'int',"months_since_policy_inception":'int',"number_of_open_complaints":'int',"number_of_policies":'int'}) 
df1.select_dtypes(np.number)
```
